```python
#Hour vs Category Popularity Heatmap
df['hour'] = df['event_time'].dt.hour

pivot = pd.crosstab(df['hour'], df['category_code'])
pivot_20 = pivot.iloc[:, :20]  # show top 20 categories

plt.figure(figsize=(14,7))
sns.heatmap(pivot_20, cmap="YlGnBu")
plt.title("Hour vs Category Popularity Heatmap")
plt.xlabel("Category")
plt.ylabel("Hour")
plt.show()
```

```python
# CATEGORY DROP-OFF ANALYSIS

# Count views per category
views_cat = df[df['event_type'] == "view"].groupby("category_code")['event_type'].count()

# Count cart events per category
carts_cat = df[df['event_type'] == "cart"].groupby("category_code")['event_type'].count()

# Combine and calculate conversion
category_conversion = (carts_cat / views_cat * 100).fillna(0).sort_values()

print("Top 5 Highest Drop-Off Categories (Lowest Conversion):")
print(category_conversion.head(5))
```

```python
# BRAND DROP-OFF ANALYSIS

# Count views per brand
views_brand = df[df['event_type'] == "view"].groupby("brand")['event_type'].count()

# Count cart events per brand
carts_brand = df[df['event_type'] == "cart"].groupby("brand")['event_type'].count()

# Combine and calculate conversion
brand_conversion = (carts_brand / views_brand * 100).fillna(0).sort_values()

print("Top 5 Highest Drop-Off Brands (Lowest Conversion):")
print(brand_conversion.head(5))
```

```python
# filter brands with at least 50 views
filtered = views_brand[views_brand > 50]
dropoff_filtered = ((carts_brand / filtered) * 100).fillna(0).sort_values()
dropoff_filtered.head(5)
```

```python
# Filter only necessary events
df_filtered = df[df['event_type'].isin(['view', 'cart', 'purchase'])]

# Group by category and event type
category_funnel = (
    df_filtered
    .groupby(['category_code', 'event_type'])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

# Calculate drop-offs
category_funnel['view_to_cart_dropoff'] = 1 - (category_funnel['cart'] / category_funnel['view'])
category_funnel['cart_to_purchase_dropoff'] = 1 - (category_funnel['purchase'] / category_funnel['cart'])

# Replace inf or NaN with 0 for safe handling
category_funnel = category_funnel.replace([np.inf, -np.inf], 0).fillna(0)

# Convert to %
category_funnel['view_to_cart_dropoff'] = (category_funnel['view_to_cart_dropoff'] * 100).round(2)
category_funnel['cart_to_purchase_dropoff'] = (category_funnel['cart_to_purchase_dropoff'] * 100).round(2)

category_funnel
```
